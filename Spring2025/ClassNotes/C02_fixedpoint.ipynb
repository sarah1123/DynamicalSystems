```python
import sympy as sym
```

```python
x = sym.Symbol('x')
equil_equation = sym.Eq(0, 4*x**2-16)
roots = sym.solve(equil_equation, x)
print(roots)
```

```python
equil_equation2 = sym.Eq(0,x-sym.cos(x))
root = sym.nsolve(equil_equation2, x, 1.0)
print(root)
```

### Symbolic Graphing

```python
from sympy.plotting import plot
p1 = plot(sym.tanh(x),(x, -4, 4), line_color='red', title='SymPy plot example')
p2 = plot(x/2,(x, -4, 4), line_color='black', title='SymPy plot example')
```
